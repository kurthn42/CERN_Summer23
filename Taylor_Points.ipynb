```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

font_title = {'family':'serif','color':'black','size':18}
font_axes = {'family':'serif','color':'black','size':16}
font_ticks = {'family':'serif','color':'black','size':16}
font_legend = {'family':'serif','size':16}
```

```python
file = pd.read_csv("Rate_vs_Distance_Linear_and_Taylor-Expanded_Fits_Compensation_Off.csv")
x = file["Taylor Fitted Point"]
x_err = file["Taylor Fitted Point Error"]
y = file["Run"]

exp = 143.25
        
fig, ax = plt.subplots(1, 1)
fig.set_size_inches(11, 8.5)
plt.errorbar(x, y, xerr = x_err, linestyle='', capsize = 4,ecolor='darkslategrey')
plt.scatter(x, y, label = 'Taylor Point',color='darkslategrey')
plt.plot(np.ones(len(y))*exp, y, linestyle = '--', label = "Average Chamber Distance: "+str(exp)+" cm",color='firebrick')
plt.title("X-Position for Taylor Expansion (Compensation OFF)", fontdict = font_title)
plt.xlabel("X-Position", loc='right', fontdict = font_axes)
plt.ylabel("Run", loc='top', fontdict = font_axes)
plt.xticks(fontsize = font_ticks['size'], fontname = font_ticks['family'])
plt.yticks(ticks = y, labels = y, fontsize = font_ticks['size'], fontname = font_ticks['family'])
plt.legend(prop = font_legend)
plt.show()
plt.close()
```

```python
file1 = pd.read_csv("Rate_vs_Distance_Linear_and_Taylor-Expanded_Fits_Compensation_Off.csv")
file2 = pd.read_csv("Rate_vs_Distance_Linear_and_Taylor-Expanded_Fits_Compensation_On.csv")
x = file1["Taylor Fitted Point"]
x = np.append(x, file2["Taylor Fitted Point"])
x_err = file1["Taylor Fitted Point Error"]
x_err = np.append(x_err, file2["Taylor Fitted Point Error"])

#exp = np.ones(len(x)) * 143.25
y = np.subtract(143.25, x)


        
fig, ax = plt.subplots(1, 1)
fig.set_size_inches(11, 8.5)
plt.errorbar(x, y, xerr = x_err, linestyle='', capsize = 4,ecolor='darkslategrey')
plt.scatter(x, y, label = 'Taylor Point',color='darkslategrey')
plt.plot(np.ones(len(y))*exp, y, linestyle = '--', label = "Average Chamber Distance: "+str(exp)+" cm",color='firebrick')
plt.title("X-Position for Taylor Expansion (Compensation OFF)", fontdict = font_title)
plt.xlabel("X-Position", loc='right', fontdict = font_axes)
plt.ylabel("Run", loc='top', fontdict = font_axes)
plt.xticks(fontsize = font_ticks['size'], fontname = font_ticks['family'])
plt.yticks(ticks = y, labels = y, fontsize = font_ticks['size'], fontname = font_ticks['family'])
plt.legend(prop = font_legend)
plt.show()
plt.close()
```
